# Linear Regression Demo

This notebook demonstrates the **linear regression** implemented from scratch:
We will:
- Generate a dataset with a linear relationship
- Train the model
- Visualize the fitted line
- Explore the effect of different hyperparameters

```python
import sys
import os

# Ensure to find our model
sys.path.append(os.path.abspath("../ml_algorithms"))

from linear_regression import LinearRegression
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Step 1: Create the Dataset
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 2 * X.flatten() + 1 + np.random.randn(50) * 2  # random noise
```

```python
plt.figure()
plt.scatter(X, y)
plt.show()
```

```python
# Create model with default parameters
model = LinearRegression()
model.fit(X, y)
```

```python
# Visualize the result

plt.figure()
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X),
         color='red', label='Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

```python
# Let's Explore parameters
model = LinearRegression(n_iterations=10000, learning_rate=0.0001)
model.fit(X, y)

plt.figure()
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X),
         color='red', label='Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
